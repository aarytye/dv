```python
import pandas as pd
names=['Area', 'Room', 'Parking', 'Warehouse', 'Elevator', 'Address', 'Price_Tomans', 'Price_USD']
rinkinys = pd.read_csv("rinkinys.csv", names=names, header=0)
nrinkinys = pd.DataFrame(rinkinys)
nrinkinys.dropna(inplace=True)
nrinkinys.head()
```

###  1

```python
import matplotlib.pyplot as plt
nrinkinys.hist(bins=19, color='green', edgecolor='black', linewidth=1.0, xlabelsize=10, ylabelsize=10, grid=False)    
plt.tight_layout(rect=(0, 0, 1.5, 1.5))  
```

###  2

```python
fig = plt.figure(figsize=(9, 7))
t = fig.suptitle('Liftas: Palyginimas tarp Kambarių skaičiaus, Ploto m2 ir Kainos USD', fontsize=14)
ax = fig.add_subplot(111, projection='3d')

xs = list(nrinkinys['Price_USD'])
ys = list(nrinkinys['Area'])
zs = list(nrinkinys['Room'])
data_points = [(x, y, z) for x, y, z in zip(xs, ys, zs)]
colors = ['green' if wt == 'Yra' else 'blue' for wt in list(nrinkinys['Elevator'])]

for data1, color in zip(data_points, colors):
    x, y, z = data1
    ax.scatter(x, y, z, alpha=0.4, c=color, edgecolors='none', s=30)

ax.set_xlabel('Kaina USD')
ax.set_ylabel('Plotas m2')
ax.set_zlabel('Kambarių skaičius')
```
