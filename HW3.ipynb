```python
import pandas as pd
names=['Area', 'Room', 'Parking', 'Warehouse', 'Elevator', 'Address', 'Price_Tomans', 'Price_USD']
rinkinys = pd.read_csv("rinkinys.csv", names=names, header=0)
rinkinys['Price_Tomans'] = rinkinys.Price_Tomans.astype(float)
nrinkinys = pd.DataFrame(rinkinys)
nrinkinys.dropna(inplace=True)
```

------------------------------------------------
- Area – plotas m^2
- Room – kambarių skaičius (1-5)
- Parking – automobilių stovėjimo aikštelė (yra/nėra)
- Warehouse – sandėlis (yra/nėra)
- Elevator – liftas (yra/nėra)
- Address – regionas, kuriame yra namas
- Price_Tomans – kaina tomanais
- Price_USD – kaina JAV doleriais
------------------------------------------------

# Vizualizacijos

```python
nrinkinys.head()
```

### Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
#https://seaborn.pydata.org/generated/seaborn.boxplot.html
plt.figure(figsize=(12, 3))
viz = sns.boxplot(x="Room", y="Warehouse", data=nrinkinys, orient='h', palette = "Set2")
plt.title('Sandėlio egzistavimas pagal kambarių skaičių',size = 15)
plt.ylabel("Sandėlys",fontsize=10)
plt.xlabel("Kambarių skaičius",fontsize=10)
plt
```

```python
nrinkinys['Elevator'].value_counts()
```

### Donut Chart

```python
colors = ['#8EB897', '#DD7596']
plt.pie(x=nrinkinys["Elevator"].value_counts().values, labels=['Yra', 'Nėra'], autopct='%.1f%%', colors=colors)
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title("Kiek procentaliai yra ir nėra liftų?")
plt.show()
```

### Scatter Plot

```python
#https://seaborn.pydata.org/generated/seaborn.scatterplot.html
plt.figure(figsize=(8, 8))
viz2 = sns.scatterplot(x="Area", y="Price_USD", data=nrinkinys, palette="deep", hue="Room")
plt.title('Kambarių skaičiaus pasiskirstymas tarp teritorijos dydžio ir kainos USD',size = 15, x=0.5, y=1.02)
plt.ylabel("Kaina USD",fontsize=12)
plt.xlabel("Teritorija m2",fontsize=12)
plt
```

### Violin Plot

```python
#https://seaborn.pydata.org/generated/seaborn.violinplot.html
#cut= --> Don’t let density extend past extreme values in the data:
plt.figure(figsize=(10, 6))
viz4 = sns.violinplot(x="Room", y="Price_USD", data=nrinkinys, palette="Set2", cut=0)
plt.title('Kaina namams pagal kambarių skaičių',size = 15, x=0.5, y=1.02)
plt.ylabel("Kaina USD",fontsize=12)
plt.xlabel("Kambarų skaičius",fontsize=12)
plt

```
